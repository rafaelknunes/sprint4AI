# # **Análise de Sentimento**

```python
### Criando subeste para análises
# #
#df = df.sample(n=10)
#
#df.shape
```

```python
import pandas as pd
from typing import List
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances

from openai import OpenAI


client = OpenAI(  
    api_key='',
    organization='org-',
    project='',
    max_retries=5
    )
    

counter = 1
def get_sentiment(text: str, model="gpt-4o-mini", **kwargs):

    global counter

    print(f"Estamos no passo: {counter}")

    # Substituir quebras de linha por espaços
    text = text.replace("\n", " ")
    

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=0, #Controla a aleatoriedade da amostra. 0 significa que a amostra é determinística
        max_tokens=5, #Número de tokens da resposta ou seja, o tamanho da resposta
        messages=[
            {
                "role": "system",
                "content": "As an AI with expertise in language and emotion analysis, your task is to analyze the sentiment of the following text, which is a complaint against a financial institution. Please focus on the customer's emotions and attitudes toward the company, taking into account the overall tone of the discussion, the language used, and the context in which words and phrases are used."
            },
            {
                "role": "user",
                "content": f"Analyze the sentiment of the following text:\n\n{text}\n\nReturn the score, from -1 to 1. Provide only the answer. For example: '-1', '0.4', etc."
            }
        ]
    )

    counter += 1

    sentiment = response.choices[0].message.content

    return sentiment

# Criando uma nova coluna para armazenar os sentimentos
df['Complaint sentiment'] = df['Consumer complaint narrative'].apply(lambda x: get_sentiment(x) if pd.notnull(x) else None)
```
